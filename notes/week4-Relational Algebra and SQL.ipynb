### 2.Join $\Join$

#### (1) natural join
1. Compute $R\times S$; 
2. equal values attributes; 
3. project unique attributes

$S1\Join R1=\pi_{\dots}(\sigma_{unique?}(S1\times R1))=$
| sid | sname   | rating | age | bid   |   day    |
|-----|---------|--------|-----|-------|----------|
| 22  | dustin  | 7      | 45  | 101   | 10/10/96 |
| 58  | rusty   | 10     | 35  | 103   | 11/12/96 |


#### (2) condition join (/ theta-join)
$R\Join_c S=\sigma_c(R\times S)$

$S1 \Join_{S1.sid<R1.sid>} R1=$
| sid | sname   | rating | age | sid | bid   |   day    |
|-----|---------|--------|-----|-----|-------|----------|
| 22  | dustin  | 7      | 35  | 58  | 103   | 11/12/96 |
| 31  | lubber  | 8      | 55  | 58  | 103   | 11/12/96 |

**Equi-Join:** condition $c$ contains only equalities (e.g. S1.sid = R1.sid)

<font color=red>Still different from natural join, since we can choose the attribute.</font>

> e.g. (1) Find names of sailors who have reserved boat #103
>
> solution 1: $\pi_{sname}((\sigma_{bid=103}Reserves)\Join Sailors)$
>
> solution 2: $\pi_{sname}(\sigma_{bid=103}(Reserves\Join Sailors))$

> e.g. (2) Find all pairs of sailors in which the older sailor has a lower rating
>
> solution 1: $\pi_{sname}((\sigma_{(S1.age<S2.age\ \wedge\ S1.rating > S2.rating)\ \vee\ (S1.age>S2.age\ \wedge\ S1.rating < S2.rating) }S1)\Join S2)$
>
> solution 2: $\pi_{sname}(\sigma_{(S1.age<S2.age\ \wedge\ S1.rating > S2.rating)\ \vee\ (S1.age>S2.age\ \wedge\ S1.rating < S2.rating) }(S1\Join S2))$

#### INSERT, UPDATE, REPLACE, DELETE
1. `INSERT`:
```SQL
# table must already exist
INSERT INTO NewEmployee
SELECT * FROM Employee;

# Without specifying fields: All columns must be specified
INSERT INTO Employee VALUES
    (DEFAULT, "A", "A's Addr", "2012-02-02", NULL, "S"),
    (DEFAULT, "B", "B's Addr", "2012-02-02", NULL, "S"),
    (DEFAULT, "C", "C's Addr", "2012-02-02", NULL, "S")
;

# Specifying fields:
INSERT INTO Employee 
    (Name, Address, DateHired, EmployeeType)
    VALUES
    (DEFAULT, "A", "A's Addr", "2012-02-02", "S"),
    (DEFAULT, "B", "B's Addr", "2012-02-02", "S"),
    (DEFAULT, "C", "C's Addr", "2012-02-02", "S")
;
```
2. `UPDATE`: Specifying a `WHERE` clause is important; Order of statements is important
```SQL
# update the whole table
UPDATE Hourly
    SET HourlyRate = HourlyRate * 1.10;

# wrong order
UPDATE Salaried
    SET AnnualSalary = AnnualSalary * 1.05
    WHERE AnnualSalary <= 10000;
UPDATE Salaried
    SET AnnualSalary = AnnualSalary * 1.05
    WHERE AnnualSalary > 10000;

# Should use CASE
UPDATE Salaried
    SET AnnualSalary = 
        CASE 
            WHEN AnnualSalary <= 10000
            THEN AnnualSalary * 1.05
            ELSE AnnualSalary * 1.10
        END;
```
3. `REPLACE`: used when you are unsure whether a record exists or not; if not exists, like INSERT. O.W. like UPDATE

4. `DELETE`: for FKs:<br/>
**ON DELETE CASCADE**: 删掉所有指向它的records<br/>
**ON DELETE RESTRICT**: 禁止我们删除该条，如果存在指向它的records
```SQL
# delete all records
DELETE FROM Employee;

# delete specified records
SELETE FROM Employee
    WHERE Name = "Grace";
```

Once a view is defined, the definition is stored in the DB (not data, but metadata-schema information); So can be used as other tables

Pros
1. Hide the query comlexity from users
2. Hide data from users (different users are set to use different views)
